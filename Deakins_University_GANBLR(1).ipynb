```python
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('iris.data', names=['sl', 'sw', 'pl', 'pw', 'target'])
label_encoder = LabelEncoder()
df['target'] = label_encoder.fit_transform(df['target'])

# this is a discrete version of adult since GANBLR requires discrete data.
# df = get_demo_data('Adult')
X_real, y_real= df.values[:,:-1], df.values[:,-1]

```

```python
model = GANBLR()
model.fit(X_real, y_real, epochs = 10)
```

```python
synthetic_data = model.sample(10000)
```

```python
# synthetic_data = synthetic_data.astype(int)
```

```python
%pip install sklearn.linear_model 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_synthetic,y_synthetic = synthetic_data[:,:-1],synthetic_data[:,-1].astype(int)
```

```python
LR = LogisticRegression(solver='liblinear')
```

```python
LR.fit(X_synthetic,y_synthetic)
```

```python
Correct_outputs_synthetic = LR.predict(X_synthetic) == y_synthetic
```

```python
Correct_outputs_synthetic.tolist().count(True)
```

```python
Correct_outputs_synthetic.sum()/X_synthetic.shape[0]
## This implies our LR model is nearly 80% accurate
```

```python
Correct_outputs_real = LR.predict(X_real) == y_real
```

```python
(Correct_outputs_real.sum())/X_real.shape[0]
## This implies that our LR model in both real data set and generated data set, which suggests that this technique of generating data actually generates data close to real is consistent
```
